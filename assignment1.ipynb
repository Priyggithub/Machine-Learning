```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sheets/Heart.csv")
```

```python
print(df.describe())
```

```python
male = len(df[df['Sex'] == 1])
female = len(df[df['Sex'] == 0])

plt.figure(figsize=(9,6))


labels = 'Male', 'Female'
sizes = [male, female]
colors = ['skyblue', 'yellowgreen']
explode = (.1,0)

plt.pie(sizes, explode=explode,labels=labels,colors=colors, autopct='%1.1f%%',shadow=True, startangle=90)

plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
labels ='chest Pain type:0','chest pain ype=1', 'chest pain type =2','chest pain type=3'
sizes = [len(df[df['Ca'] ==0]),len(df[df['Ca'] ==1]),len(df[df['Ca'] ==2]),len(df[df['Ca'] ==3])]
colors = ['skyblue', 'yellowgreen','orange','gold']
explode =(0,0,0,0)

plt.pie(sizes,explode=explode, labels=labels, colors=colors, autopct='1%.1f%%',shadow=True,startangle=180)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
labels = 'fasting blood sugar<120mg/dl','fasting sugar>120mg/dl'
sizes = [len(df[df['Fbs'] == 0]),len(df[df['Ca'] == 1])]
colors = ['skyblue','yellowgreen']
explode = (0.1,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True,startangle=180)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
labels='NO','Yes'
sizes=[len(df[df['ExAng'] == 0]),len(df[df['ExAng'] == 0])]
colors=['skyblue','yellowgreen']
explode=(0.1,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=100)
plt.axis('equal')
plt.show()
```
